# Significancy of Correlation between Goals Difference and Clasification

```python
import pandas as pd
import seaborn as sns
```

```python
df5ligas=pd.read_excel('data\\Clasificacion_5_Grandes_Ligas_2021_2022.xlsx')
```

```python
df5ligas=df5ligas[['RL', 'DG']]
```

```python
sns.pairplot(df5ligas)
```

```python
corr_matrix = df5ligas.corr()
heatmap = sns.heatmap(corr_matrix, annot=True)
heatmap.set(ylim=(0,len(corr_matrix)))
heatmap.set(xlim=(0,len(corr_matrix)))
```
